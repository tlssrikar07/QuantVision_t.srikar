```python
import yfinance as yf
import plotly.graph_objects as go
import plotly.io as pio


pio.renderers.default = "colab"
```

```python
ticker_symbol = "MSFT"
ticker = yf.Ticker(ticker_symbol)

# Last 6 months, 1-day interval
historical_data = ticker.history(period="6mo", interval="1d")

historical_data.head()
```

```python
fig = go.Figure(
    data=[
        go.Candlestick(
            x=historical_data.index,
            open=historical_data["Open"],
            high=historical_data["High"],
            low=historical_data["Low"],
            close=historical_data["Close"]
        )
    ]
)

fig.update_layout(
    title="Microsoft (MSFT) Candlestick Chart – Last 6 Months",
    xaxis_title="Date",
    yaxis_title="Price (USD)",
    xaxis_rangeslider_visible=True
)

fig.show()
```

### Candlestick Pattern Analysis

**Stock Chosen:** Microsoft (MSFT)

**Date of Pattern:**19 September 2025

**Pattern Name:** Bullish Continuation  

**Price Movement After Pattern:**The prices still went up confirming uptrend
